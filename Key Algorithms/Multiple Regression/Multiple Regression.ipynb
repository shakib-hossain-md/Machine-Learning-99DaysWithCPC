```python
# plot Weight vs CO2 with prediction value
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("G:/Machine Learning/w3 School/Multiple Regression/data.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X, y)

newdata= pd.DataFrame({'Weight':[2300], 'Volume': [1300]})
predictionCO2 = regr.predict(newdata)

# plot weight vs CO2
plt.scatter(df['Weight'], y, color='blue', label='Actual Data: Weight vs CO2')

# plot the prediction point
plt.scatter(newdata['Weight'], predictionCO2, color='red', label='Predicted CO2', marker='X', s=100)

#add labels and title
plt.xlabel('Weight(g)')
plt.ylabel('CO2 Emissions')
plt.title('Weight vs CO2 Emissions with prediction value')

plt.legend()
plt.show()
```

```python
# regression line for both Weight and Volume A 3D surface plot

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D


# Load the dataset
df = pd.read_csv("G:/Machine Learning/w3 School/Multiple Regression/data.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X, y)

new_data = pd.DataFrame({'Weight': [2300], 'Volume': [1300]})
predictedCO2 = regr.predict(new_data)


# Create a grid of values for the 3D surface
weight_range = np.linspace(df['Weight'].min(), df['Weight'].max(), 100)
volume_range = np.linspace(df['Volume'].min(), df['Volume'].max(), 100)
weight_grid, volume_grid = np.meshgrid(weight_range, volume_range)

# Predict CO2 emissions for the grid
CO2_predicted = regr.predict(pd.DataFrame({'Weight': weight_grid.ravel(), 'Volume': volume_grid.ravel()}))
CO2_predicted = CO2_predicted.reshape(weight_grid.shape)

 #Create a 3D plot for the multiple regression surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(weight_grid, volume_grid, CO2_predicted, color='lightblue', alpha=0.6)

# Plot the actual data points
ax.scatter(df['Weight'], df['Volume'], y, color='blue', label='Actual Data', s=50)

# Plot the predicted point
ax.scatter(new_data['Weight'], new_data['Volume'], predictedCO2, color='red', label='Predicted CO2 (Weight=2300, Volume=1300)', marker='X', s=100)

# Set labels for axes
ax.set_xlabel('Weight')
ax.set_ylabel('Volume')
ax.set_zlabel('CO2 Emissions')

# Show the plot
plt.show()
```

```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df = pd.read_csv("G:/Machine Learning/w3 School/Multiple Regression/data.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

# Create the linear regression model and fit the data
regr = linear_model.LinearRegression()
regr.fit(X, y)

# Predict the CO2 emission for a specific car
new_data = pd.DataFrame({'Weight': [2300], 'Volume': [1300]})
predictedCO2 = regr.predict(new_data)

# Create a range of values for plotting the regression line
weight_range = np.linspace(df['Weight'].min(), df['Weight'].max(), 100)
volume_range = np.linspace(df['Volume'].min(), df['Volume'].max(), 100)

# Generate predictions using the regression model
predicted_CO2_weight = regr.predict(pd.DataFrame({'Weight': weight_range, 'Volume': np.full_like(weight_range, 1300)}))
predicted_CO2_volume = regr.predict(pd.DataFrame({'Weight': np.full_like(volume_range, 2300), 'Volume': volume_range}))

# Plot 1: Weight vs CO2 with regression line
plt.scatter(df['Weight'], y, color='blue', label='Actual Data: Weight vs CO2', alpha=0.6)
plt.plot(weight_range, predicted_CO2_weight, color='orange', label='Regression Line (Volume=1300)', linewidth=2)

# Plot the predicted point
plt.scatter(new_data['Weight'], predictedCO2, color='red', label='Predicted CO2 (Weight=2300, Volume=1300)', marker='X', s=100)

# Add labels and title
plt.xlabel('Weight (g)')
plt.ylabel('CO2 Emissions')
plt.title('Weight vs CO2 Emissions with Regression Line')

# Show legend
plt.legend()

# Display the plot
plt.show()

# Plot 2: Volume vs CO2 with regression line
plt.scatter(df['Volume'], y, color='green', label='Actual Data: Volume vs CO2', alpha=0.6)
plt.plot(volume_range, predicted_CO2_volume, color='orange', label='Regression Line (Weight=2300)', linewidth=2)

# Plot the predicted point
plt.scatter(new_data['Volume'], predictedCO2, color='red', label='Predicted CO2 (Weight=2300, Volume=1300)', marker='X', s=100)

# Add labels and title
plt.xlabel('Volume (ccm)')
plt.ylabel('CO2 Emissions')
plt.title('Volume vs CO2 Emissions with Regression Line')

# Show legend
plt.legend()

# Display the plot
plt.show()

# Combined Plot: Weight and Volume vs CO2
plt.scatter(df['Weight'], y, color='blue', label='Actual Data: Weight vs CO2', alpha=0.6)
plt.scatter(df['Volume'], y, color='green', label='Actual Data: Volume vs CO2', alpha=0.6)

# Plot the predicted points for both Weight and Volume
plt.scatter(new_data['Weight'], predictedCO2, color='red', label='Predicted CO2 (Weight=2300, Volume=1300)', marker='X', s=100)

# Add labels and title
plt.xlabel('Weight (g) and Volume (ccm)')
plt.ylabel('CO2 Emissions')
plt.title('Weight and Volume vs CO2 Emissions with Prediction')

# Show legend
plt.legend()

# Display the plot
plt.show()
```
