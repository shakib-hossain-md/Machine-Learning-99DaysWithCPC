```python
# Removing Rows
# Ex: Remove rows with a NULL value in the 'Date' column.
import pandas as pd
df=pd.read_csv('D:\\programming\\Machine Learning 99DaysWithCPC\\Python Basics\\Pandas\\exercise_data.csv')
df['Date']=pd.to_datetime(df['Date'])
df.dropna(subset=['Date'],inplace=True)
print(df.to_string())
```

```python
# Fixing Wrong Data
# Replacing Value
# Ex: Set "Duration"= 45 in row 7
import pandas as pd
df=pd.read_csv('D:\\programming\\Machine Learning 99DaysWithCPC\\Python Basics\\Pandas\\exercise_data.csv')
df.loc[7,'Duration']=45
print(df.to_string())
```

```python
# Ex: Loop through all values in the "Duration" column.
# If the value is higher than 120, set it to 120:
import pandas as pd
df=pd.read_csv('D:\\programming\\Machine Learning 99DaysWithCPC\\Python Basics\\Pandas\\exercise_data.csv')
for x in df.index:
    if df.loc[x,"Duration"]>115:
        df.loc[x,"Duration"]=115
print(df.to_string())
```

```python
# Removing Rows
# Ex: Delete rows where "Duration" is greater than 115.
import pandas as pd
df=pd.read_csv('D:\\programming\\Machine Learning 99DaysWithCPC\\Python Basics\\Pandas\\exercise_data.csv')
for x in df.index:
    if df.loc[x,"Duration"]>115:
        df.drop(x,inplace=True)
print(df.to_string())
```

```python
# Discovering Duplicates
# Ex:Returns True for every row that is a duplicate, otherwise False:
import pandas as pd
df=pd.read_csv('D:\\programming\\Machine Learning 99DaysWithCPC\\Python Basics\\Pandas\\exercise_data.csv')
print(df.duplicated())
```
