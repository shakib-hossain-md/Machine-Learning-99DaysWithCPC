```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
numeric_data = tips.select_dtypes(include=[float, int])
print(numeric_data.corr())
```

```python
tc = numeric_data.corr()
sns.heatmap(tc, cmap='Blues', annot=True)
```
