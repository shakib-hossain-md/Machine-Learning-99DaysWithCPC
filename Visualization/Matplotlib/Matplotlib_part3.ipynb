## Plot Appearance

---



```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
fig =  plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue',linewidth=1,alpha=0.5,linestyle='--')
# You can also add hexadecimal color code.
```

## Marker

---



```python
fig =  plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue',linewidth=1,alpha=0.5,marker='o')
```

```python
# Marker Size
fig =  plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue',linewidth=1,alpha=0.5,marker='o',markersize=20)
```

```python
# Marker face color
fig =  plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue',linewidth=1,alpha=0.5,marker='o',markersize=20,markerfacecolor='yellow')
```

```python
# Mrker edge width
fig =  plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue',linewidth=1,alpha=0.5,marker='o',markersize=20,markerfacecolor='yellow',
markeredgewidth=3)
```

```python
# Mrker edge widtcolor
fig =  plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue',linewidth=1,alpha=0.5,marker='o',markersize=20,markerfacecolor='yellow',
markeredgewidth=3,markeredgecolor='green')
```

## Line Limit

---



```python
fig =  plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue',linewidth=1,linestyle='--')
ax.set_xlim(0,1)
ax.set_ylim(0,2)
```
