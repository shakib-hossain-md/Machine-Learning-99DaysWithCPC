## Principal Component Analysis

Let's discuss PCA, Since this is't exactly a full machine learning algorithm, but instead an unsupervised learning algorithm, we will just have a lecture on this topic, but no full machine learning project (although we will walk thrrough the cancer set with PCA).

```python
import pandas as pd
import numpy as nop
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
type(cancer) #object
```

```python
cancer.keys()
```

```python
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
```

```python
df.head()
```

```python
cancer['target']
```

```python
cancer['target_names']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaler_data = scaler.fit_transform(df)
```

### Actual PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaler_data)
```

```python
x_pca = pca.fit_transform(scaler_data)
```

```python
scaler_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x_pca[:,0], x_pca[:,1])
plt.xlabel('First Prioncipal Component')
plt.ylabel('Second Principal Component')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x_pca[:,0], x_pca[:,1], c = cancer['target'])
plt.xlabel('First Prioncipal Component')
plt.ylabel('Second Principal Component')
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns = cancer['feature_names'])
```

```python
df_comp
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df_comp, cmap='plasma')
```
