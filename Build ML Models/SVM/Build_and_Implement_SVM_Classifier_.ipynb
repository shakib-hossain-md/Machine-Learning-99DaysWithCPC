### Build Support Vector Machine Classifier

---



```python
from threading import Condition
class SVM_classifier():

  def __init__(self, learning_rate, no_of_iteration, lambda_parameter):
    self.learning_rate = learning_rate
    self.no_of_iteration = no_of_iteration
    self.lambda_parameter = lambda_parameter

  def fit(self, X, Y):
    self.m, self.n = X.shape

    self.w = np.zeros(self.n)
    self.b = 0
    self.X = X
    self.Y = Y

    for i in range(self.no_of_iteration):
      self.update_weights()

  def update_weights(self):
    y_label = np.where(self.Y <= 0, -1, 1)

    for index, x_i in enumerate(self.X):
      Condition = y_label[index] * (np.dot(x_i, self.w) - self.b) >=1

      if (Condition == True):
        dw = 2 * self.lambda_parameter * self.w
        db = 0

      else:
        dw = 2 * self.lambda_parameter * self.w - np.dot(x_i, y_label[index])
        db = y_label[index]

      self.w = self.w - self.learning_rate * dw
      self.b = self.b - self.learning_rate * db

  def predict(self, X):
    output = np.dot(X, self.w) - self.b
    predicted_labels = np.sign(output)
    y_hat = np.where(predicted_labels <= -1, 0, 1)

    return y_hat
```

### Importing the Dependencies

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```


---

### Data Collection and Pre-*Processing*

```python
# loading the data from csv file to pandas DataFrame
diabetes_data = pd.read_csv('/content/diabetes.csv')
```

```python
# print the first 5 rows of the dataframe
diabetes_data.head()
```

```python
diabetes_data['Outcome'].value_counts()
```

```python
# separating the features and target
features = diabetes_data.drop(columns = 'Outcome', axis=1)
target = diabetes_data['Outcome']
```

```python
features
```

```python
target
```

### Data Standardization

---



```python
scaler = StandardScaler()
```

```python
scaler.fit(features)
```

```python
Standardized_data = scaler.transform(features)
```

```python
Standardized_data
```

```python
features = Standardized_data
target = diabetes_data['Outcome']
```

### Train Test Split

---



```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=2)
```

```python
features.shape, X_train.shape, X_test.shape
```

### Train the Model

---



```python
svm_classifier = SVM_classifier(learning_rate=0.001, no_of_iteration=1000, lambda_parameter=0.01)
```

```python
# training the svm classifier with training data
svm_classifier.fit(X_train, y_train)
```

### Model Evaluation

---



Accuracy Score

```python
# accuracy score on training data
X_train_prediction = svm_classifier.predict(X_train)
```

```python
training_data_accuracy = accuracy_score(y_train, X_train_prediction)
```

```python
training_data_accuracy
```

```python
# accuracy score on testing data
X_test_prediction = svm_classifier.predict(X_test)
```

```python
test_data_accuracy = accuracy_score(y_test, X_test_prediction)
```

```python
test_data_accuracy
```
