```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
diabetes_datset = pd.read_csv('/content/diabetes.csv')
```

```python
diabetes_datset.head()
```

```python
diabetes_datset.shape
```

```python
# Separating features and Target
X = diabetes_datset.drop(columns='Outcome', axis=1)
Y = diabetes_datset['Outcome']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape,X_test.shape)
```

Model Training

KNN Classifier in Sklearn

```python
classifier = KNeighborsClassifier(p=1)
```

```python
classifier.fit(X_train,Y_train)
```

Model Evaluation

```python
y_pred = classifier.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test, y_pred)
```

```python
print(accuracy*100)
```
